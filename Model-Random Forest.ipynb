```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_dataset.csv")
df.head()
```

# Label Categorization

```python
category_amt = 10
limiters = np.arange(1/category_amt,1,1/category_amt)
limiters
```

```python
df['gdp_per_capita_category'] = np.nan
percentiles = df.groupby('Year')['gdp_per_capita'].quantile(limiters).unstack()
percentiles
```

```python
def categorize(row):
    year = row['Year']
    val = row['gdp_per_capita']
    
    q = np.zeros(len(limiters))
    
    for i in range(len(limiters)):
        q[i] = percentiles.loc[year, limiters[i]]
    
    pos = 0
    
    for i in q:
        if val <= i:
            break
        else:
            pos+=1
    return pos            
```

```python
df['gdp_per_capita_category'] = df.apply(categorize, axis=1)
```

```python
df.drop(['gdp_per_capita'], axis=1, inplace=True) # Unused
df.drop(['Entity', 'Year'], axis = 1, inplace=True) # Irrelevant

df.head()
```

# Model Creation

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x = df.drop(['gdp_per_capita_category'], axis = 1)
y = df['gdp_per_capita_category']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state=314)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=314)
model.fit(x_train, y_train)
prediction = model.predict(x_test)

print(f"Accuracy = {metrics.accuracy_score(prediction, y_test)}")
```

```python
precision = metrics.precision_score(y_test, prediction, average=None)
    
confusionMatrix = metrics.confusion_matrix(y_test, prediction)
```

```python
category_amt = 10

fig, ax =plt.subplots(nrows=1, ncols=2, figsize=(14,5))
plt.subplots_adjust(hspace=.3)

ax[0].bar(np.arange(0,category_amt,1), precision)
for i, height in enumerate(precision):
    ax[0].text(i, height-0.06, str(int(1000*round(height, 3))/10), ha='center', color='white', fontsize='small')

ax[0].set_title(f"Precision percentage for each category")
ax[0].set_xticks(np.arange(0,category_amt, 1))
ax[0].set_ylabel("Percentage")
ax[0].set_xlabel("Category")
ax[0].plot()
fig.suptitle("Random Forest Classifier", fontsize='x-large', weight='bold')


figure = metrics.ConfusionMatrixDisplay(confusionMatrix)
ax[1].set_title(f"Confussion Matrix")
figure.plot(ax=ax[1])
```
